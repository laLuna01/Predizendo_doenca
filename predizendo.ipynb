```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Blood_samples_dataset_balanced_MODIFICADO.csv")
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
x = df.drop("Disease", axis=1)
y = df["Disease"]
```

```python
sabedor = DecisionTreeClassifier()
sabedor.fit(x, y)
```

```python
plot_tree(sabedor)
```

```python
paciente = x.sample()
paciente
```

```python
sabedor.predict(paciente)
```

```python
accuracy_score(y, sabedor.predict(x))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
sabidinho = DecisionTreeClassifier()
sabidinho.fit(x_train, y_train)
```

```python
accuracy_score(y_test, sabidinho.predict(x_test))
```
